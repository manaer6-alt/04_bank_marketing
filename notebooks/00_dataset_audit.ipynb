## Stage 0 — Dataset audit and task framing

Project: `04_bank_marketing`

Dataset: Bank Marketing Dataset

### Dataset size

- Rows: 11162
- Columns: 17

### Target

Target column: `deposit`

Positive class: `yes`

Class distribution:

- `no`: 5873 / 52.6%
- `yes`: 5289 / 47.4%

There is no severe class imbalance, but this is still a business classification problem where precision/recall trade-off matters.

### Problem type

Binary classification.

### Initial business framing

Predict whether a client will subscribe to a term deposit before the final outcome of the marketing contact is known.

This framing is useful for campaign prioritization: the model can help decide which clients are more promising to contact.

### Numeric columns

Original numeric columns:

- `age`
- `balance`
- `day`
- `duration`
- `campaign`
- `pdays`
- `previous`

Candidate numeric columns:

- `age`
- `balance`
- `day`
- `campaign`
- `pdays`
- `previous`

### Categorical columns

Candidate categorical columns:

- `job`
- `marital`
- `education`
- `default`
- `housing`
- `loan`
- `contact`
- `month`
- `poutcome`

### Missing values

No explicit NaN values were found.

However, several categorical columns contain `unknown` values:

- `poutcome`: 8326
- `contact`: 2346
- `education`: 497
- `job`: 70

For the initial pipeline, `unknown` will be treated as a valid category and encoded with OneHotEncoder.

### Duplicates

Duplicate rows: 0

### Date-like columns

The dataset contains:

- `day`
- `month`

There is no full date, year, timestamp, or campaign id. Because of that, a fully reliable chronological split is not possible from the available columns alone.

Initial validation plan:

- use stratified train/test split;
- reserve test set for final evaluation only;
- use cross-validation on train during model comparison and tuning.

### Leakage risks

`duration` is a high-risk leakage feature for the pre-contact business framing.

Reason: call duration is only known during or after the call. If the model is supposed to prioritize clients before contact, `duration` is unavailable at prediction time.

Decision:

- exclude `duration` from candidate features.

`pdays`, `previous`, and `poutcome` describe previous campaign history. They are not automatically leakage if this information is available before the current campaign contact.

Decision:

- keep `pdays`, `previous`, and `poutcome` as candidate features for now.

### Initial metrics

Primary metrics:

- PR-AUC
- F1
- precision
- recall

Secondary metrics:

- ROC-AUC
- accuracy as auxiliary only

Accuracy should not be the main metric because business decisions depend on the cost of false positives and false negatives.

### Main business trade-off

False positive:

- bank contacts a client who will not subscribe;
- cost: wasted call/sales effort, possible customer irritation.

False negative:

- bank does not prioritize a client who would have subscribed;
- cost: missed business opportunity.

Threshold tuning must be done only on train/CV, never on the final test set.
